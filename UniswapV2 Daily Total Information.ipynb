This is a simple example script to show you how to pull UniswapV2 summary information with one call

```python
# import packages
import pandas as pd
import requests
from datetime import datetime
```

```python
# make subgraph graphql call
url = 'https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2'

query = """
{
    uniswapDayDatas(first: 1000, orderBy: date, orderDirection: desc) {
        dailyVolumeETH
        dailyVolumeUSD
        dailyVolumeUntracked
        date
        totalLiquidityETH
        totalLiquidityUSD
        totalVolumeETH
        totalVolumeUSD
        txCount
    }
}"""
response = requests.post(url, json={'query': query})
```

```python
# convert to dataframe
df = pd.json_normalize(response.json()['data']['uniswapDayDatas'])
```

```python
# turn timestamp into datetime
df['date'] = pd.to_datetime(df['date'], unit='s')
df
```

```python
# download df as csv
df.to_csv('uniswapV2_daily_summary.csv', index=False)
```
